```python
import yfinance as yf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = yf.download(tickers='TSLA',start='2020-01-01')
```

```python
df['Log_Returns'] = np.log(df['Close'] / df['Close'].shift(1))
df.dropna(inplace=True)
```

```python
rolling_window = 21  # 21 trading days ~ 1 month
df['Rolling_Volatility'] = df['Log_Returns'].rolling(window=rolling_window).std()
df['Squared_Returns'] = df['Log_Returns'] ** 2
df['Vol_of_Vol'] = df['Rolling_Volatility'].rolling(window=rolling_window).std()

# Plot the rolling volatility
plt.figure(figsize=(12, 6))
plt.plot(df['Rolling_Volatility'], label=f'{rolling_window}-Day Rolling Volatility', color='blue')
plt.xlabel('Date')
plt.ylabel('Volatility')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['Squared_Returns'], label='Squared Returns', color='orange')
plt.title('Volatility Clustering')
plt.xlabel('Date')
plt.ylabel('Squared Returns')
plt.legend()
plt.show()
```

```python
# Plot vol of vol
plt.figure(figsize=(12, 6))
plt.plot(df['Vol_of_Vol'], label='Volatility of Volatility', color='green')
plt.xlabel('Date')
plt.ylabel('Vol of Vol')
plt.legend()
plt.show()
```

## Volatility signature plot

Use the equation
$$
\sigma^2(\tau) = \sigma^2_{r} \left[ 1 + 2 \sum^\tau_{u=1} \left(1-\frac{u}{\tau} \right) C_{r}(u) \right]
$$
where
$r =$ log return

$\sigma_{r} = \mathrm{Var}(r)$

$C_{r}(\tau) = \mathrm{Cov}(r_{t},r_{t+\tau})/\sigma^2_{r}$

```python
sigma_r2 = np.var(df['Log_Returns'])  # Variance of returns

def autocorrelation(log_returns, lag):
    n = len(log_returns)
    r = log_returns - np.mean(log_returns)
    cov = np.sum(r[:-lag] * r[lag:]) / n
    return cov / sigma_r2


def volatility_signature(log_returns, max_tau):
    sigmas = []
    for tau in range(1, max_tau + 1):
        sum_term = 0
        for u in range(1, tau + 1):
            weight = 1 - (u / tau)
            sum_term += weight * autocorrelation(log_returns, u)
        
        sigma_tau2 = sigma_r2 * (1 + 2 * sum_term)
        sigmas.append(np.sqrt(sigma_tau2))
    return sigmas

max_tau = 30
volatility_signature_plot = volatility_signature(df['Log_Returns'].values, max_tau)

plt.figure(figsize=(10, 6))
plt.plot(range(1, max_tau + 1), volatility_signature_plot, marker='o', color='blue', label='Volatility Signature')
plt.xlabel('Time Interval (Days)')
plt.ylabel('Volatility')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```
